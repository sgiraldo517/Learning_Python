Matplotlib is one of the Python packages that allow us to do **Data Visualization**. Data visualization is veru important for data analysis. It help us to:

### Basic plots with Matplotlib


`````python
# First we will need to import pyplot (a subpackage from Matplotlib)
import matplotlib.pyplot as plt

year = [1950, 1970, 1990, 2012]
pop = [2.519, 3.692, 5.263, 6.972]

# To make a line plot: 
plt.plot(year, pop)
    ## Note: the first argument corresponds to the horizontal axis (x-axis) and the sendo to the vertical one (y-axis) 
    
# To display a plot you will need the show argument: 
plt.show()
`````

When you have a time scale along the horizontal axis, the line plot is your friend. But in many other cases, when you're trying to assess if there's a correlation between two variables, for example, the scatter plot is the better choice. Below is an example of how to build a scatter plot.

`````python
# To draw a scatter plot we will use the following code (This type of plot will only display the individual data points):
plt.scatter(year, pop)
plt.show()

# Something usefull to add to the scatter plot is the size of the dots to better represent the data shown
import numpy as np
np_pop = np.array(pop)
np_pop = np_pop * 30

plt.scatter(year, pop, s=np_pop)
plt.show()
    ## Note: sometimes it is also useful to convert the axis into a logarithmic scale. use the following code: 
    ## plt.xscale('log')
`````

### Dictionaries
Allow us to store information in an alternative way to list so that it is easier to identify the corresponding values. To build a dictionary you will use **{}** and the data will be organized as follows: 

`dictionary = {"keyword1":value, "keyword2":value, "keyword3":value]`

`````python
# Example of a dictionary
world = {"afghanistan":30.55, "albania": 2.77, "algeria":39.21}

# To access the value of an element you pass the key of the corresponding value in []
world["albania"]
`````

`````python
# Comparison between index method vs dictionaries

countries = ['spain', 'france', 'germany', 'norway']
capitals = ['madrid', 'paris', 'berlin', 'oslo']

## INDEX METHOD
# Get index of 'germany': ind_ger
ind_ger = countries.index("germany")

# Use ind_ger to print out capital of Germany
print(capitals[ind_ger])

europe = {'spain':'madrid', 'france': 'paris', 'germany':'berlin', 'norway': 'oslo'}

## DICTIONARY METHOD
# Print out the keys in europe
print(europe.keys())

# Print out value that belongs to key 'norway'
print(europe['norway'])
`````

#### Characteristics of the keys
- The keys in a dictionary need to be **unique**: If the keys are duplicated, only the last value specified will be included taken into account
- Keys have to be immutable objects: the content of the immutable objects can not be changed after they are created
     >**Note**  int, strings, floats and booleans are immutable objects. Lists are not. This means that lists can not be set as keys for a dictionary 

`````python
# Add an element to a dictionary 
world["sealand"] = 0.000027
print(world)

# To check if a key is included in a dictionary
print('sealand' in world)

# Update a value
world["sealand"] = 0.000028
print(world)

# To delete an element from a dictionary 
del(world["sealand"])
print(world)
`````

#### Lists vs. Dictionaries 
|                List                 |               Dictionary            |
| ----------------------------------- | ----------------------------------- |
| Select, update, and remove with [ ] | Select, update, and remove with { } |
| Indexed by range of numbers         | Indexed by unique keys              |
| Collection of values - orders matters, for selecting entire subsets | Lookup table with unique keys |

#### Dictionary of dictionaries

`````python
# Dictionary of dictionaries
europe = { 'spain': { 'capital':'madrid', 'population':46.77 },
           'france': { 'capital':'paris', 'population':66.03 },
           'germany': { 'capital':'berlin', 'population':80.62 },
           'norway': { 'capital':'oslo', 'population':5.084 } }


# Print out the capital of France
print(europe['france']['capital'])

# Create sub-dictionary data
data = {'capital': 'rome', 'population':59.83}

# Add data to europe under key 'italy'
europe["italy"] = data

# Print europe
print(europe)
`````

### Pandas
High level data maniputalion tool built on the Numpy package. It allows to store the tabular data in an element called a **Dataframe**

#### Creating Dataframes from dictionaries
Having the following dictionary: 
```
dict = { 
    "country": ["Brazil", "Russia", "India", "China", "South Africa"],
    "capital": ["Brasilia", "Moscow", "New Delhi", "Beijing", "Pretoria"],
    "area": [8.516, 17.10, 3.286, 9.597, 1.221],
    "population": [200.4, 143.5, 1252, 1347, 52.98] }
```
- Keys will be the **column** labels
- Values will be the **data**, column by column

`````python
# Create the dict
dict = { 
    "country": ["Brazil", "Russia", "India", "China", "South Africa"],
    "capital": ["Brasilia", "Moscow", "New Delhi", "Beijing", "Pretoria"],
    "area": [8.516, 17.10, 3.286, 9.597, 1.221],
    "population": [200.4, 143.5, 1252, 1347, 52.98] }

# Import Pandas
import pandas as pd

# Convert the dict into a pandas Dataframe
brics = pd.DataFrame(dict)
print(brics)
print(type(brics))

# To update the authomatic indexes assigned by pandas
brics.index = ['BR','RU','IN','CH','SA']
print(brics)
`````

`````python
# Pre-defined lists
names = ['United States', 'Australia', 'Japan', 'India', 'Russia', 'Morocco', 'Egypt']
dr =  [True, False, False, False, True, True, True]
cpc = [809, 731, 588, 18, 200, 70, 45]

# Import pandas as pd
import pandas as pd

# Create dictionary my_dict with three key:value pairs: my_dict
my_dict = { 
    "country": names,
    "drives_right": dr,
    "cars_per_cap": cpc
}

# Build a DataFrame cars from my_dict: cars
cars = pd.DataFrame(my_dict) 

# Print cars
print(cars)

# Definition of row_labels
row_labels = ['US', 'AUS', 'JPN', 'IN', 'RU', 'MOR', 'EG']

# Specify row labels of cars
cars.index = row_labels

# Print cars again
print(cars)
`````

#### Import data from an external source
Frequently, we will need to import data from CSV files. To do it we will need the following code: 

`database = pd.read_csv("path/to/file.csv")`

  >**Note:** if necesarry you can specify the column that correspond to the indexes by adding the parameter: `index_col= 0`

#### Index and Select data from a Dataframe
There are 2 methods to select data from dataframes with pandas:
1. **Square brackets:**

    To select a column from a dataframe with square brackets you only need to write the name of the Dataframe and the name of the column you want to select in square brackets `dataframe['column']`
      >**Note:** when we use a single [], the column will be saves as a pandas series. To save the data as a dataframe we should use [[ ]] `dataframe[['column1','column2']]`
    
    To select rows we nees to use the slice method `dataframe[1:4]`. The previous code will select rows from the 2nd to the 5th row.
    
    *Limitations: the square brackets method will only allow us to select either columns or rows, not a combiantion of both*

2. **Loc and Iloc:** allow us to create both columns and rows of a dataframe
    
    - **loc:**  is label-based, which means that you have to specify rows and columns based on their row and column labels. 
        
        Example: 
        
        - One row and one column `cars.loc['IN', 'cars_per_cap']`
        - Multiple rows and one column `cars.loc[['IN', 'RU'], 'cars_per_cap']`
        - Multiple rows and multiple columns `cars.loc[['IN', 'RU'], ['cars_per_cap', 'country']]` 
        - Select all rows and one column `cars.loc[:, 'country']`
        - Select all rows and multiple columns `cars.loc[:, ['country','drives_right']]`
    
    - **iloc:** is integer index based, so you have to specify rows and columns by their integer index.
    
        Example: 
        
        - One row and one column `cars.iloc[3, 0]`
        - Multiple rows and one column `cars.iloc[[3, 4], 0]`
        - Multiple rows and multiple columns `cars.iloc[[3, 4], [0, 1]]`
        - Select all rows and one column `cars.iloc[:, 1]`
        - Select all rows and multiple columns `cars.iloc[:, [1, 2]]`
    
    >**Note:** To print a single column and all rows as a Dataframe use: `cars.iloc[:,[2]]`

### Comparison Operators
Operator that can tell how 2 Python values relate, and result in a boolean. 
>**Note:** Python can not tell how 2 objects with different data types relate. *Exceptions: Float and Integers, **NympyArrys** and Integers*

#### Comparators

| Comparator | Meaning |
| ------ | ---- |
| < | Strictly less than|  
| <= | Less than or equal |
| > | Strictly greater than |
| == | Equal |
| != | Not equal |




`````python
# To check if two Python values, or variables, are equal you can use ==. To check for inequality, you need !=

# Comparison of booleans
print(True == False)

# Comparison of integers
print((-5*15) != 75)

# Comparison of strings
print("pyscript" == "PyScript")

# Compare a boolean with an integer
print(True == 1)
`````

`````python
# Greater and less than

# Comparison of integers
x = -3 * 6
print(x >= -10)

# Comparison of strings
y = "test"
print("test" <= y)

# Comparison of booleans
print(True > False)
`````

`````python
# Compare arrays

# Create arrays
import numpy as np
my_house = np.array([18.0, 20.0, 10.75, 9.50])
your_house = np.array([14.0, 24.0, 14.25, 9.0])

# my_house greater than or equal to 18
print(my_house >= 18)

# my_house less than your_house
print (my_house <= your_house)
`````

### Boolean Operations 
The 3 most common operators are:
- **and:** Takes 2 booleans and returns True only if both the booleans themselves are True
- **or:** At least one of the booleans used should be True
- **not:** It simply negates the boolean value you use it on

`````python
# and, or, not

# Define variables
my_kitchen = 18.0
your_kitchen = 14.0

# my_kitchen bigger than 10 and smaller than 18?
print(my_kitchen > 10 and my_kitchen < 18)

# my_kitchen smaller than 14 or bigger than 17?
print(my_kitchen < 14 or my_kitchen > 17)

# Double my_kitchen smaller than triple your_kitchen?
print(my_kitchen*2 < your_kitchen*3)
`````

#### Boolean Operators in Numpy arrays
For Numpy arrays we need to use the following methods as the regular boolean operators will generate an error. This methods are performed element-wise: 
- **logical_and()**
- **logical_or()**
- **logical_not()**

`````python
# Boolean operators with NumPy

# Create arrays
import numpy as np
my_house = np.array([18.0, 20.0, 10.75, 9.50])
your_house = np.array([14.0, 24.0, 14.25, 9.0])

# my_house greater than 18.5 or smaller than 10
print(np.logical_or(my_house > 18.5, my_house < 10))

# Both my_house and your_house smaller than 11
print(np.logical_and(my_house <11, your_house < 11))
`````

### Conditional Statements: if, elif else

```
if condition1 :
    expression
elif condition2 :
    expression
else :
    expression
```

- The **if** conditional will run a piece of code whenever a condition is **True**.

>**Note:** the expression segment of the if structure can contain more than a single piece of code to be executed

- To specify another possible condition an piece of code to be executed we will use **elif**.
- To execute a code when **none** of the specified conditions are **true** we will use **else**.

>**Note:** As soon as Python bumps into a condition that is true, it executes the corresponding code and then stop running. 

`````python
area = 10.0
if(area < 9) :
    print("small")
elif(area < 12) :
    print("medium")
else :
    print("large")
`````

`````python
# Define variables
room = "bed"
area = 14.0

# if-elif-else construct for room
if room == "kit" :
    print("looking around in the kitchen.")
elif room == "bed":
    print("looking around in the bedroom.")
else :
    print("looking around elsewhere.")

# if-elif-else construct for area
if area > 15 :
    print("big place!")
elif area > 10 :
    print("medium size, nice!")
else :
    print("pretty small.")
`````

### Filtering pandas DataFrames
To filter data in a Pandas Dataframe we will need to follow this steps:
1. We will need to extract as a Series the column by we want to filter. 
2. Perform a comparison on the recently created column.
3. Use the column to index the DataFrame

`````python
# Import cars data
import pandas as pd
cars = pd.read_csv('Datasets/cars.csv', index_col = 0)

# Extract drives_right column as Series: dr
dr = cars['drives_right']

# Use dr to subset cars: sel
sel = cars[dr]

# Print sel
print(sel)
`````

The code in the previous example worked fine, but you actually unnecessarily created a new variable dr. You can achieve the same result without this intermediate variable. Put the code that computes dr straight into the square brackets that select observations from cars.

`````python
# Import cars data
import pandas as pd
cars = pd.read_csv('Datasets/cars.csv', index_col = 0)

# Convert code to a one-liner
sel = cars[cars['drives_right']]

# Print sel
print(sel)
`````

You can also use `np.logical_and()`, `np.logical_or()` and `np.logical_not()` on Pandas Series to do more advanced filtering operations. Please see the following example:

`````python
# Import cars data
import pandas as pd
cars = pd.read_csv('Datasets/cars.csv', index_col = 0)

# Import numpy, you'll need this
import numpy as np

# Create medium: observations with cars_per_cap between 100 and 500
cpc = cars['cars_per_cap']
between = np.logical_and(cpc > 100, cpc < 500)
medium = cars[between]

# Print medium
print(medium)
`````

### While loop
The while loop is like a repeated if statement. The code is executed over and over again, as long as the condition is `True`
```
while condition :
    expression
````

`````python
# Example 1 
# The following code will only have 3 printouts
x = 1
while x < 4 :
    print(x)
    x = x + 1
    
print("-----------------")

# Example 2
# Initialize offset
offset = 8

# Code the while loop
while offset != 0 :
    print("correcting...")
    offset = offset - 1
    print(offset)
    
print("-----------------")
    
# Example 3: add an if - else into the while
# Initialize offset
offset = -6

# Code the while loop
while offset != 0 :
    print("correcting...")
    if offset > 0 :
      offset = offset - 1
    else : 
      offset = offset + 1
    print(offset)
`````

### for loop
The structure of the for loop is the following: 
```
for variable in sequeance : 
    expression
```
The code above can be read as: *"for each variable, in a sequence, execute the expressions"*   

`````python
# Example 1 
fam = [1.73, 1.68, 1.71, 1.89]
for index, height in enumerate(fam) :
    print("index " + str(index) + ": " + str(height))

print("---------------")

    # Enumerate function: Enumerate() method adds a counter to an iterable and returns it in a form of enumerating object. This enumerated object can then be used directly for loops or converted into a list of tuples using the list() function.
        # Syntax: enumerate(iterable, start=0) 
            # iterable: any object that supports iteration
            # Start:  the index value from which the counter is to be started, by default it is 0 
            
fam = [1.73, 1.68, 1.71, 1.89]
print (list(enumerate(fam, 0)))
print (list(enumerate(fam, 2)))

print("---------------")

# Example 2: we can also iterate over every character of a string
for c in "family":
    print(c.capitalize())
print("---------------")

#Example 3: Loop over list of lists
# house list of lists
house = [["hallway", 11.25], 
         ["kitchen", 18.0], 
         ["living room", 20.0], 
         ["bedroom", 10.75], 
         ["bathroom", 9.50]]
         
# Build a for loop from scratch
for room,area in house : 
    print('the ' + room + ' is ' + str(area) + ' sqm')
`````
